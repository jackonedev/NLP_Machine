```python
# PROGRAMA PRINCIPAL
if sum([twitter, facebook, instagram]) != 1:
    print("Elija un solo filtro")
    exit()

# Librerias de Python
import pickle, sys, os, datetime
import pandas as pd
import ipyaggrid

# APP Reputaci√≥n Digital
from app.main.data_feed import main as leer_csv
from app.time_series import TimeSeries
from app.tokenizadores import tokenizador_i
from app.m4_similarity import M4_similarity


# Librer√≠as Reputaci√≥n Digital
from app.main.data_feed import main as leer_csv
from tools.feed import procesar_file_csv, data_info
from tools.facebook import clean_facebook_comments
from tools.intagram import clean_instagram_comments
```

```python
nombre, archivo = procesar_file_csv(archivo_lectura)

df = leer_csv(archivo, verbose=True)

if not twitter:
    if facebook:
        batch = clean_facebook_comments(df.content.to_list())
    elif instagram:
        batch = clean_instagram_comments(df.content.to_list())
    df = pd.DataFrame(batch, columns=["content"])

first_8 = "- " + "\n- ".join([t.replace("\n","") for t in df.head(8).content.to_list()])
print(f"Primeros 8 registros:\n{first_8}")
```

```python
try:
    path = TimeSeries.main(archivo_lectura, conservar_duplicados=True) ## default: eliminar repetidos
    with open(path, "rb") as f:
        df = pickle.load(f)
        df = df.reset_index(drop=True)
except:
    # print("Error en el procesamiento de TimeSeries")
    # print('Comprobar la existencia de "@timestamp" o "date" dentro de las columnas')
    print("Continuando ejecuci√≥n...\n")

dftk = tokenizador_i.main_df(df.copy())

df["content"] = dftk["content"]
dftk = dftk.drop(columns=["content"])

df = pd.concat([df, dftk], axis=1)
df.name = nombre



print()
# display(df.head(8))
first_8 = "- " + "\n- ".join([t.replace("\n","") for t in df.head(8).content.to_list()])
print(f"Primeros 8 registros:\n{first_8}")
data_info(df)
```

```python
## FILTRO PARA DIFERENCIAR PUBLICACIONES VS COMENTARIOS
```

```python
## IMPLEMTACION DEL MODELO M4
df = df.drop_duplicates(subset=["content"])
df = M4_similarity.main_df(df, topicos, max_workers=8)
df.name = nombre        
```

```python
# Umbrales:
umbral_probabilidad = 0.3
umbral_distancia = 0.5

# Filtro de probabilidad
low_score_mask = pd.DataFrame(df.score_similarity_i.to_list()).iloc[:, 0] < umbral_probabilidad

# Muestra para corregir las categorizaciones
df_correccion = df[low_score_mask]
# Muestra para an√°lisis
df = df[~low_score_mask]

# DataFrame auxiliares
df_scores = pd.DataFrame(df.score_similarity_i.to_list())
df_labels = pd.DataFrame(df.similarity_i.to_list())
mask_2da_etiqueta = (1- (df_scores.iloc[:, 0] - df_scores.iloc[:, 1]) / (df_scores.iloc[:, 0] + df_scores.iloc[:, 1])) > umbral_distancia


df["max_2_similarity_i"] = df_labels[mask_2da_etiqueta].iloc[:, 1]
df.max_2_similarity_i = df.max_2_similarity_i.fillna("otros")
```

```python
## NO EJECUTAR ESTA CELDA HASTA QUE ESTEN IMPLEMENTADOS LOS FILTROS DE LOW_SCORE Y SEGUNDA_ETIQUETA

## DESCARGA DE ARCHIVOS
## project_root/Z_{nombre}/backup.pkl
## project_root/Z_{nombre}/porcentajes.txt
## project_root/Z_{nombre}/{archivo}


output_dir = os.getcwd()
output_dir = os.path.join(output_dir, f"Z_{nombre}")
os.makedirs(output_dir, exist_ok=True)


with open(os.path.join(output_dir, "backup-analisis.pkl"), "wb") as f:
    df.name = nombre
    pickle.dump(df, f)

with open(os.path.join(output_dir, "backup-correccion.pkl"), "wb") as f:
    df_correccion.name = nombre
    pickle.dump(df_correccion, f)


df.to_csv(os.path.join(output_dir, archivo), index=False)


porcentajes = (df.max_similarity_i.value_counts()/df.max_similarity_i.value_counts().sum() *100).round(1).astype(str) + " %"
with open(os.path.join(output_dir, "porcentajes.txt"), "w") as f:
    f.write(str(porcentajes))

porcentajes_2 = (df.max_2_similarity_i.value_counts()/df.max_2_similarity_i.value_counts().sum() *100).round(1).astype(str) + " %"
with open(os.path.join(output_dir, "porcentajes_2.txt"), "w") as f:
    f.write(str(porcentajes_2))


```
