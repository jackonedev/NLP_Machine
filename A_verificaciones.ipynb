```python
import numpy as np
import pandas as pd
import ipyaggrid
```

```python
###  USER INPUT  ###
archivo_lectura = 'rd-fb-ilegalidad-rocio.csv'
###  USER INPUT  ###
```

```python
df = pd.read_csv(archivo_lectura)
df.emojis = df.emojis.replace("[]",np.nan)
df = df.loc[:, df.any(axis=0)]
features = ["content","sentiment_i","emotions_6_max_label","emotions_26_max_label"]
df = df[features]
df.head(4)
```

```python
column_defs = [{'field': c} for c in df.columns]
column_defs[0] = column_defs[0] | {"width":800}


grid_options = {
    'columnDefs' : column_defs,
    'enableColResize': True
}
```

```python
g = ipyaggrid.Grid(
    grid_data=df,
    theme='ag-theme-balham-dark',
    sync_grid=True,
    show_toggle_edit=True,
    sync_on_edit=True,
    grid_options=grid_options,
    export_mode='auto',
    export_to_df=True
)
```

```python
display(g)
```

```python
df_edit = g.grid_data_out['grid']
df_edit
```

```python
df_edit.to_csv(f"CORRE_{archivo_lectura}", index_col=0)
```
